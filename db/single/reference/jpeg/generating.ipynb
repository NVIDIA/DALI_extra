```python
from PIL import Image

img = Image.open('../../jpeg/134/site-1534685_1280.jpg')
```

```python
display(img)
```

```python
import numpy as np

a = np.asarray(img)

np.save('site-1534685_1280_uint8', a)
np.save('site-1534685_1280_uint16', np.uint16(a * 257))
np.save('site-1534685_1280_float', np.float32(a / 255))
```

```python
img = img.crop((20, 5, 1000, 800))
display(img)
```

```python
a = np.asarray(img)

np.save('site-1534685_1280_roi_uint8', np.uint8(a))
np.save('site-1534685_1280_roi_uint16', np.uint16(a * 257))
np.save('site-1534685_1280_roi_float', np.float32(a / 255))
```
